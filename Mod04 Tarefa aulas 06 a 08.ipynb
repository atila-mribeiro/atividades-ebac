####  1) Lendo de APIs
Vimos em aula como carregar dados públicos do governo através de um API (*Application Programming Interface*). No exemplo de aula, baixamos os dados de pedidos de verificação de limites (PVL) realizados por estados, e selecionamos apenas aqueles referentes ao estado de São Paulo.

1. Repita os mesmos passos feitos em aula, mas selecione os PVLs realizados por municípios no estado do Rio de Janeiro.
2. Quais são os três *status* das solicitações mais frequentes na base? Quais são suas frequências?
3. Construa uma nova variável que contenha o ano do **status**. Observe que ```data_status``` vem como tipo *object* no **DataFrame**. Dica: você pode usar o método ```.str``` para transformar o tipo da variável em string, em seguida um método como [**slice()**](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.slice.html) ou [**split()**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.split.html).
4. Indique a frequência de cada ano do campo construído no item (3).

````python
# 1)
# baixando a base de dados e salvando em um dataframe
url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Munic%C3%ADpio'
r = requests.get(url)

data_json = r.json()
df = pd.DataFrame(data_json['items'])

# Verificando que foram selecionados apenas os municípios do estado do RJ
df['uf'].unique()
````

````python
# 2) Contando quantas vezes os valores da variável 'status' aparece
df['status'].value_counts()
````

Para a variável 'status' os valores que mais se repetem são: deferido, arquivado e regularizado, com 96, 53 e 26 saídas respectivamente.

````python
# 3) Seu código aqui

# Transformando a variável em string
df['data_status'] = df['data_status'].astype(str)

# Extraindo o ano do status usando o método split()
df['ano_status'] = df['data_status'].str.split('/').str[2]

# Imprimindo a nova variável 'ano_status'
print(df['ano_status'])
````

````python
# 3)Verificando a frequência de cada ano

df['ano_status'].value_counts()
````

####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui


def consultar_pvl(uf, tipo_interessado):
    # Montando a URL da consulta
    url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'
    
    # Fazendo a requisição HTTP
    response = requests.get(url)

    # Convertendo a resposta em formato JSON
    data_json = response.json()
        
    # Criando o DataFrame a partir dos dados JSON
    df = pd.DataFrame(data_json['items'])
        
    # Retornando o DataFrame
    return df

#Selecionando os argumentos escolhidos
uf = 'MG'
tipo_interessado = 'Estado'

#Chamando a função
dados_df = consultar_pvl(uf, tipo_interessado)

# Mostrando o resultado
dados_df
````

````python
# 2) Verificando a contagem de valores da variável 'status'
dados_df['status'].value_counts()
````

Está registadro apenas 1 solicitação para o estado de MG com o status 'Arquivado por decurso de prazo'.

````python
# 3) 
#Selecionando os argumentos escolhidos
uf = 'BA'
tipo_interessado = 'Município'

#Chamando a função
dados_df = consultar_pvl(uf, tipo_interessado)

# Filtrando as solicitações deferidas para o estado da Bahia
solicitacoes_bahia_deferidas = (dados_df['status'] == 'Deferido')

# Imprimindo o município com mais solicitações deferidas
dados_df[solicitacoes_bahia_deferidas]['interessado'].value_counts()
````

O município da Bahia com mais solicitações Deferidas é o Luís Eduardo Magalhães

````python
# 4) Seu código aqui

#Selecionando os argumentos escolhidos
uf = 'BA'
tipo_interessado = 'Estado'

#Chamando a função
dados_df = consultar_pvl(uf, tipo_interessado)

#Salvando o arquivo em csv
dados_df.to_csv('pvl_bahia.csv')
````
